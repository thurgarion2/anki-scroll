```python
from train.wikipedia_index_search import recall_metric, load_wikipedia_dataset, recall_feedback, CustomInstructionProposer
from anki_scroll.service.website_query import WikipediaIndex
from anki_scroll.llms import grok_fast, oss_120
import dspy

examples = load_wikipedia_dataset()

index = WikipediaIndex()
index.set_lm(grok_fast)
```

```python
eval = dspy.Evaluate(
    devset=examples, 
    metric=recall_metric, 
    display_progress=True, 
    display_table=True,
    num_threads=10)
```

```python
#Average Metric: 8.48 / 13 (65.2%): 100%|██████████| 13/13 [01:07<00:00,  5.23s/it]
#2025/12/06 14:12:52 INFO dspy.evaluate.evaluate: Average Metric: 8.476190476190476 / 13 (65.2%)
```

```python
gepa = dspy.GEPA(
    metric=recall_feedback,
    max_metric_calls=150,
    num_threads=7,
    track_stats=True,
    use_merge=False,
    instruction_proposer=CustomInstructionProposer(),
    reflection_lm=oss_120)

index_optimised = gepa.compile(index, trainset=examples, valset=examples)
```

```python
from train.gepa_report import generate_gepa_report
from pathlib import Path
html = generate_gepa_report(index_optimised.detailed_results)
Path("./wikipedia_index_report.html").write_text(html)
```

```python
from train.config import data_folder
index_optimised.save(data_folder/"wikipedia_index.json")
```
